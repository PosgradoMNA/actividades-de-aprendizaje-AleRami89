# 2.Modelo PCA

## 2.1 Selección de variables para el PCA 

PCA solo acepta variables numéricas.
Por lo que generamos una copia sin las columnas categóricas del DataFrame llamada numeric_data.

```python
numeric_data = ndf.drop(
    columns=["X2", "X3", "X4", "X6", "X7", "X8", "X9", "X10", "X11", "Y"]
)
numeric_data.head()
```

## 2.2 Normalización de datos usando "preprocessing.scale"



No es necesario escalar con otro método ya que este es el que ayuda a la normaliazión de datos. 

```python
scaled_data = preprocessing.scale(numeric_data)
scaled_data
```

## 2.3 Modelo PCA

```python
pca = PCA()
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)
pca_data
```

Porcentaje de varianza por componente

```python
per_var = np.round(pca.explained_variance_ratio_ * 100, decimals=1)
labels = ["PC" + str(x) for x in range(1, len(per_var) + 1)]

plt.bar(x=range(1, len(per_var) + 1), height=per_var, tick_label=labels)
plt.ylabel("Porcentaje de la varianza")
plt.xlabel("Componente Principal")
plt.title("PCA Histograma")
plt.show()
```

Resumen de PCA

```python
pcsSummary_df = pd.DataFrame(
    {
        "Standard deviation": np.sqrt(pca.explained_variance_),
        "Proportion of variance": pca.explained_variance_ratio_,
        "Cumulative proportion": np.cumsum(pca.explained_variance_ratio_),
    }
)
pcsSummary_df = pcsSummary_df.T
pcsSummary_df
```

Los componentes de PC1 a PC8 tienen el 96.9% de varianza acumulada siendo estos los mas representativos

Transponer el data frame para una mejor lectura y para ver como cada varible original (x1-x23) contribuye a los pesos de cada componente (PC1-PC14).

```python
pca_components_df = pd.DataFrame(
    pca.components_.T, columns=labels, index=numeric_data.columns
)
pca_components_df
```

# 3.Scatter Plot
